### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,6,8,10]

if not (isinstance(x,(list,tuple)) and isinstance(y,(list,tuple))):
    print("Error")
elif len(x)!=len(y):
    print("Error")
else:
    plt.scatter(x,y,color='red')
    for i,j in zip(x,y):
        plt.annotate(f'({i},{j})', xy=(i,j),textcoords="offset points",xytext=(5,-10))
    plt.title("Scatter Plot with Annotations")
    plt.xlabel("X axis")
    plt.ylabel("Y axis")
    plt.grid(True)
```
